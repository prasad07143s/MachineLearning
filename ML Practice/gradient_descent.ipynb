```python
# Importing the necessary modules

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Implementing gradient descent algorithm

x = np.array([1, 2, 3, 4, 5, 6, 7])
y = np.array([val ** 2 for val in x])

def gradient_descent(x, y):
    iterations, learning_rate, n = 8186, 0.01, len(x)
    slope = intercept = 0
    for i in range(iterations):
        y_predicted = slope * x + intercept
        mse = (1 / n) * sum([val ** 2 for val in (y - y_predicted)])
        d_slope = (2 / n) * sum(-x * (y - y_predicted))
        d_intercept = (2 / n) * sum(-(y - y_predicted))
        slope -= learning_rate * d_slope
        intercept -= learning_rate * d_intercept
        print(slope, intercept, mse)
        plt.plot(x, y_predicted, color = "green")
    plt.plot(x, y_predicted, color = "black", lw = 2)

gradient_descent(x, y)
plt.show()
```

```python
# Calculating the slope and intercept using sklearn

from sklearn import linear_model

X = x.reshape(-1, 1)

model = linear_model.LinearRegression()

model.fit(X, y)

y_predicted = model.predict(X)

print(model.score(X, y))

print(model.coef_, model.intercept_)
```
