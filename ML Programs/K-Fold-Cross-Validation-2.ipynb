```python
# Importing the necessary modules

import numpy as np
import pandas as pd
from sklearn import linear_model, svm, ensemble, model_selection, datasets
```

```python
# Loading the dataset

dataset = datasets.load_digits()
print(dir(dataset))
```

```python
# Creating the dataframe

df = pd.DataFrame(dataset.data, columns = dataset.feature_names)
df['target'] = dataset.target
df.head()
```

```python
# Splitting the data into 'k' folds

X = df.drop('target', axis = 1).values
y = df['target'].values

kf = model_selection.StratifiedKFold(n_splits = 10)
```

```python
# Evaluating the model performance

lg = linear_model.LogisticRegression()
SVM = svm.SVC()
rf = ensemble.RandomForestClassifier()

lg_scores = []
svm_scores = []
rf_scores = svm_scores.copy()

def train_model(model, X_train, X_test, y_train, y_test):
    model.fit(X_train, y_train)
    return model.score(X_test, y_test)

for train_index, test_index in kf.split(X, y):
    X_train, X_test, y_train, y_test = X[train_index], X[test_index], y[train_index], y[test_index]
    lg_scores.append(train_model(lg, X_train, X_test, y_train, y_test))
    svm_scores.append(train_model(SVM, X_train, X_test, y_train, y_test))
    rf_scores.append(train_model(rf, X_train, X_test, y_train, y_test))
def avg(li):
    return sum(li) / len(li)
print(avg(lg_scores))
print(avg(svm_scores))
print(avg(rf_scores))
```

```python
# Conclusion : SVM performs better in classification
```
