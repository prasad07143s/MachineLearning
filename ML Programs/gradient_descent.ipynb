```python
### Author : Prasad Meesala
# Importing the necessary moduels

import numpy as np
import matplotlib.pyplot as plt
```

```python
# Drawing line of linear regression using gradient descent algorithm

def gradient_descent(x, y):
    iterations, learning_rate, n = 10000, 0.01, len(x)
    slope = intercept = 0
    for i in range(iterations):
        y_predicted = slope * x + intercept
        mse = (1 / n) * sum(val ** 2 for val in (y - y_predicted))
        d_slope = (2 / n) * sum(-x * (y - y_predicted))
        d_intercept = (2 / n) * sum(-(y - y_predicted))
        slope -= learning_rate * d_slope
        intercept -= learning_rate * d_intercept
        print(f"Slope : {slope} Intercept : {intercept} mse : {mse}")
        plt.plot(x, y_predicted, color = "red")
    plt.plot(x, y_predicted, color = "blue")
    plt.show()
    
x = np.array([1, 2, 3, 4, 5])
y = np.array([1, 4, 9, 16, 25])
gradient_descent(x, y)
```

```python
# Drawing the line of linear regression using sklearn

import pandas as pd
from sklearn import linear_model

X = pd.DataFrame(np.array([1, 2, 3, 4, 5])).values
y = np.array([val ** 2 for val in x])

reg = linear_model.LinearRegression()
reg.fit(X, y)
y_predicted = reg.predict(X)

plt.plot(X.reshape(-1), y_predicted, lw = 3)
print(f"Slope : {reg.coef_} Intercept : {reg.intercept_}")
plt.show()
```
