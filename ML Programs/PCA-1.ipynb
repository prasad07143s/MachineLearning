```python
# Importing the necessary modules

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import cluster, model_selection, preprocessing, decomposition
```

```python
# Importing the dataset

df = pd.read_csv("wine-clustering.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.corr()
```

```python
# Visualzing the correlations between the attributes

plt.figure(figsize = (15, 7))
sns.heatmap(df.corr(), annot = True)
plt.show()
```

```python
# Scaling the features (standardization)

scaler = preprocessing.StandardScaler()

new_df = pd.DataFrame(scaler.fit_transform(df), columns = df.columns)
new_df.head()
```

```python
# Reducing the number of dimensions using PCA

pcs = decomposition.PCA(n_components = 2)

df = pd.DataFrame(pcs.fit_transform(new_df), columns = pcs.get_feature_names_out())

print(df.shape)
df.head()
```

```python
print(dir(pcs))
```

```python
print(pcs.n_components)
```

```python
print(pcs.explained_variance_)
print(pcs.n_features_)
```

```python
# Scaling the principal components

scaler = preprocessing.MinMaxScaler()
df = pd.DataFrame((scaler.fit_transform(df)), columns = ['pca0', 'pca1'])
df.head()
```

```python
# Visualizing the data
    
plt.scatter(df['pca0'], df['pca1'])
plt.show()
```

```python
# Choosing the optimal value of 'k' in k-means

sse = []

for i in range(1, 10):
    model = cluster.KMeans(n_clusters = i)
    model.fit(df)
    sse.append(model.inertia_)

plt.plot(range(1, 10), sse)
plt.show()
```

```python
# Fitting the model and training it with k = 3

model = cluster.KMeans(n_clusters = 3)

labels = model.fit_predict(df)
print(labels)
```

```python
# Assigning labels

df['target'] = labels
df.head()
```

```python
# Visualizing the clusters

df0 = df[df['target'] == 0]
df1 = df[df['target'] == 1]
df2 = df[df['target'] == 2]

plt.scatter(df0['pca0'], df0['pca1'])
plt.scatter(df1['pca0'], df1['pca1'])
plt.scatter(df2['pca0'], df2['pca1'])
plt.show()
```
