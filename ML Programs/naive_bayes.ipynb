```python
# Importing the necessary modules

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets, model_selection, naive_bayes, metrics
```

```python
# Collecting the data

data = datasets.load_wine()
print(dir(data))

df = pd.DataFrame(data.data, columns = data.feature_names)
df['target'] = data.target
df['target_names'] = df['target'].apply(lambda x : data.target_names[x])
df.info()
df.head()
```

```python
# Visualizing the data

plt.figure(figsize = (20, 10))
sns.heatmap(df.corr(), annot = True)
plt.show()

plt.hist(bins = 10, x = df['alcohol'])
plt.show()

plt.hist(bins = 10, x = df['flavanoids'])
plt.show()

df.describe()
```

```python
# Splitting the data into training and testing data

X = df.drop(['target', 'target_names'], axis = 1).values
y = df['target'].values

X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size = 0.25)
```

```python
# Training the model

model1 = naive_bayes.GaussianNB()

model2 = naive_bayes.MultinomialNB()

model1.fit(X_train, y_train)
model2.fit(X_train, y_train)
```

```python
# Testing the model and predicting the outcome

y_predicted = model1.predict(X_test)
print(y_test, y_predicted)
```

```python
# Checking the accuracy of the model

print(model1.score(X_train, y_train))
print(model1.score(X_test, y_test))

print(model2.score(X_train, y_train))
print(model2.score(X_test, y_test))

## Determining the accuracy through confusion matrix

sns.heatmap(metrics.confusion_matrix(y_test, y_predicted), annot = True)
plt.show()

sns.heatmap(metrics.confusion_matrix(y_test, model2.predict(X_test)), annot = True)
plt.show()
```
