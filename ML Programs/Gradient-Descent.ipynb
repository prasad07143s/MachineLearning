#### Implementing gradient descent algorithm

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gradient_descent(x, y):
    n, learning_rate = 1000, 0.1
    slope = intercept = 0
    for i in range(n):
        y_predicted = slope * x + intercept
        mse = (1 / n) * sum(val ** 2 for val in y - y_predicted)
        mse_slope = (-2 / n) * sum((y - y_predicted) * x)
        mse_intercept = (-2 / n) * sum(y - y_predicted)
        slope -= learning_rate * mse_slope
        intercept -= learning_rate * mse_intercept
        print(slope, intercept, mse)
        plt.plot(x, y_predicted, color = "blue")
    plt.plot(x, y_predicted, color = "red")
    plt.show()
        
x = np.array([1, 2, 3, 4, 5])
y = np.array([1, 4, 9, 16, 25])
gradient_descent(x, y)
```

```python
from sklearn import linear_model
import pandas as pd

model = linear_model.LinearRegression()

x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

model.fit(pd.DataFrame(x), y)
print(model.coef_, model.intercept_)
```
